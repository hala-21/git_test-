```python
import pandas as pd # importing laibrary used for data manipulation and analysis.
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data' # link to data set
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']   # adding column's names

df = pd.read_csv(url, names=column_names,
                         na_values = "?", comment='\t',
                         sep=" ", skipinitialspace=True)
df                 # "df" refer to  dataframe which is  is a 2-dimensional labeled data structure provided by the pandas library
```

```python
#Summing  missing values along columns
df.isna().sum()
```

```python
# finding concise summary of a DataFrame.
df.info()
```

```python
# finding first 3 raws of dataframe
df.head(3)
```

```python
#finding size f=of data frame (raws * columns)
df.shape
```

```python
#droping all null values and assign it to new data frame called "df2"
df2=df.dropna()
df2
```

```python
#finding sum of null values in df2
df2.isna().sum()
```

```python
#import sklearn library to do math operations on the data frame
from sklearn.preprocessing import MinMaxScaler
# assign minn to max method to the  variable called "mms" which we will use to do the normalaization process to increase accuracy
mms = MinMaxScaler()  # note that MinMaxScaler does n't know which frame to work on untill the fit _transform is called and given the dataframe as an argument
#do the fit & transform operations to fit the values from the scalling process to the real dataframe values then  transforming the real data to the scaled one then assign the normalaized data to a new data frame called df3
df3 = mms.fit_transform(df2)
df3
```

```python
#or we can do the scalling &fitting and transforming all in one step :)
from sklearn.preprocessing import MinMaxScaler
df3 = MinMaxScaler().fit_transform(df2)
df3
```

```python
#since the new data frame "df3" is now consist of ranges of parameters
#let's make anew data frame "df4"consist of same values from "df3"  not as ranges but as values like in "df2"
df4=pd.DataFrame(df3,columns=df2.columns)
df4
```

```python
#now let's put the first column in a variable called "MPG"
MPG = df4[['MPG']]
MPG
```

```python
#then let's remove it (drop it ) from dataframe "df4" and assign the values to a new data frame called "df5"
df5 = df4.drop(labels=['MPG'],axis=1)
df5
```
